```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt("Layout_9.txt")
x = data[:, 0]
y = data[:, 1]
```

```python
a = np.linspace(-1000, 1000, 20)
b = np.linspace(-1000, 1000, 20)

detector_pos = [(i, j) for i in a for j in b]
a = np.array(detector_pos)[:, 0]
b = np.array(detector_pos)[:, 1]
```

```python
plt.figure(figsize = [6, 6])
plt.scatter(x, y, color = "blue")
plt.scatter(a, b, color = "green", alpha = .3)
plt.xlim((-1500, 1500))
plt.ylim((-1500, 1500))
```
